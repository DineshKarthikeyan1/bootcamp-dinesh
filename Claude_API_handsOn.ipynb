```python
from google.colab import userdata

api_key = userdata.get('dinkey')
```

```python
import anthropic


# connecting to the api
client = anthropic.Anthropic(api_key=userdata.get('dinkey'))


# send a message
message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=256,
    messages=[
        {"role": "user", "content": "What is 2 + 2? Answer in one sentence."}
    ]
)
```

```python
print(message.content[0])
print()
print(message)
print()
print("\n--- Token Usage ---")
print(f"Input Tokens: {message.usage.input_tokens}")
print(f"Output Tokens: {message.usage.output_tokens}")
```

## Multi-Turn Convos

```python
# Conversation history is going to be a list
conversation = []

def chat(user_message):
  # Add the new user message to history
  conversation.append({"role": "user", "content": user_message})

  # send full history everytime
  response = client.messages.create(
      model="claude-haiku-4-5-20251001",
      max_tokens=256,
      messages=conversation
  )

  # extract text response from the received JSON
  reply = response.content[0].text

  # append claude reply to memory
  conversation.append({"role": "assistant", "content": reply})

  # token stats
  print(f"Claude: {reply}")
  print(f"Input Tokens: {response.usage.input_tokens}")
  print(f"Output Tokens: {response.usage.output_tokens}")
```

```python
chat("My name is Alex and I'm learning the Anthropic API.")
chat("What's my name?")
chat("What am I learning?")
```
